```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img,cmap='gray')
```

```python
edges = cv2.Canny(img,threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(img,threshold1=220,threshold2=220)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF MEDIAN VALUE WHICHEVER IS GREATER
lower = int(max(0,0.7*med_val))
# HIGHER THRESHOLD TO EITHER 130% OR MAX 255 OF MEDIAN VALUE WHICHEVER IS SMALLER
upper= int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
